```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("../train.csv")
test = pd.read_csv("../test.csv")
```

```python
train.head()
```

```python
not_concerned_columns = ["PassengerId", "Name", "Ticket", "Fare", "Cabin", "Embarked"]

train = train.drop(columns=not_concerned_columns)
test = test.drop(columns=not_concerned_columns)
```

```python
test.head()
```

```python
train.head()
```

```python
train.describe()
```

```python
train.isnull().sum()
```

```python
train["Age"] = train["Age"].fillna(train["Age"].median())
```

```python
train.isnull().sum()
```

```python
test["Age"] = test["Age"].fillna(test["Age"].median())
```

```python
test.isnull().sum()
```

```python
train.head()
```

```python
train["Sex"].unique()
```

```python
train["Sex"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(train["Sex"].unique())
```

```python
train["Sex"] = le.transform(train["Sex"])
test["Sex"] = le.transform(test["Sex"])
```

```python
train.head()
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
```

```python
train["Age"].values.reshape(-1, 1)
```

```python
train["Age"] = ss.fit_transform(train["Age"].values.reshape(-1, 1))
test["Age"] = ss.fit_transform(test["Age"].values.reshape(-1, 1))
```

```python
train.head()
```

```python
test.head()
```

```python
from sklearn.model_selection import train_test_split

train_X = train.drop(columns=['Survived'])
train_Y = train.Survived

(train_X, test_X, train_Y, test_Y) = train_test_split(train_X, train_Y, test_size = 0.3, random_state = 666)
```

```python
from sklearn.metrics import (roc_curve, auc, accuracy_score)
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=0)
clf = clf.fit(train_X, train_Y)
pred = clf.predict(test_X)

fpr, tpr, thresholds = roc_curve(test_Y, pred, pos_label=1)
auc(fpr, tpr)
accuracy_score(pred, test_Y)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

features = train_X.columns
importances = clf.feature_importances_
indices = np.argsort(importances)

plt.figure(figsize=(6,6))
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), features[indices])
plt.show()
```

```python
my_prediction = clf.predict(test.values)
my_prediction.shape
```

```python
test2 = pd.read_csv("../test.csv")
PassengerId = np.array(test2.PassengerId).astype(int)
PassengerId
```

```python
my_solution = pd.DataFrame(my_prediction, PassengerId, columns = ["Survived"])
my_solution.to_csv("my_random_forest.csv", index_label = ["PassengerId"])
```
