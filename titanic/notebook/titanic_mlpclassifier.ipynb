```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("../train.csv")
test = pd.read_csv("../test.csv")
```

```python
not_concerned_columns = ["PassengerId", "Name", "Ticket", "Fare", "Cabin", "Embarked"]

train_data = train.drop(columns=not_concerned_columns)
test_data = test.drop(columns=not_concerned_columns)
```

```python
train_data = train_data.replace("male", 0).replace("female", 1)
test_data = test_data.replace("male", 0).replace("female", 1)
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
train_data.Age = train_data.Age.fillna(train_data.Age.median())
train_data.describe()
```

```python
test_data.Age = test_data.Age.fillna(test_data.Age.median())
test_data.describe()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
```

```python
X = train_data.drop(columns = ["Survived"]).values
Y = train_data.Survived.values

train_X, train_test_X, train_Y, train_test_Y = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
clf = MLPClassifier()

clf.fit(train_X, train_Y)

predict = clf.predict(train_test_X)
predict
print(classification_report(train_test_Y, predict))
```

```python
test_X = test_data.values

clf_submit = MLPClassifier()

clf_submit.fit(X, Y)

predict_submit = clf_submit.predict(test_X)
predict_submit
```

```python
PassengerId = np.array(test.PassengerId).astype(int)
PassengerId
```

```python
my_solution = pd.DataFrame(predict_submit, PassengerId, columns = ["Survived"])
my_solution.to_csv("../my_mlpclassifier.csv", index_label = ["PassengerId"])
```
